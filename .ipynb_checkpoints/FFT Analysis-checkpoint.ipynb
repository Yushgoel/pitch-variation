```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import librosa
import librosa.display
from scipy.stats import median_abs_deviation

import IPython.display as ipd
```

```python
sr = 44000

audio = 'hello.wav'
x, _ = librosa.load(audio, sr=sr)

audio2 = 'hello_stretch.wav'
x2, _ = librosa.load(audio2, sr=sr)

audio3 = 'hello_long.wav'
x3, _ = librosa.load(audio3, sr=sr)

audio4 = "hello_repeat.wav"
x4, _ = librosa.load(audio4, sr=sr)
```

```python
print("Simple Hello")
ipd.Audio(x, rate = sr)
```

```python
print("Hello with different intonation")
ipd.Audio(x2, rate = sr)
```

```python
print("Simple Hello + Silence")
ipd.Audio(x3, rate = sr)
```

```python
print("Repeating Simple Hello for Consistency Check")
ipd.Audio(x4, rate = sr)
```

```python
plt.figure(figsize=(14, 5))
librosa.display.waveplot(x, sr = sr)
```

```python
x_fft = np.fft.fft(x)
x2_fft = np.fft.fft(x2)
x3_fft = np.fft.fft(x3)
x4_fft = np.fft.fft(x4)
```

```python
plt.figure(figsize=(14, 5))
plt.plot(np.abs(x_fft))
```

```python
x_fft_p1 = x_fft[0: int(len(x_fft) / 2)]
x2_fft_p1 = x2_fft[0: int(len(x2_fft) / 2)]
x3_fft_p1 = x3_fft[0: int(len(x3_fft) / 2)]
x4_fft_p1 = x4_fft[0: int(len(x4_fft) / 2)]
```

```python
x_abs = np.abs(x_fft_p1)
x2_abs = np.abs(x2_fft_p1)
x3_abs = np.abs(x3_fft_p1)
x4_abs = np.abs(x4_fft_p1)
```

```python
np.std(x_abs), np.std(x2_abs), np.std(x3_abs), np.std(x4_abs)
```

Simple std isn't consistent across recording. Also, realized doesn't make sense. Simple std is not comparing frequencies, but magnitude (confidence) of frequencies.

```python
# Picking top 10,000 frequencies (maybe too large a number since it will pick up background noise). 

# Will look at reducing it when analyzing sentences and different accents / backgrounds.

n = 10000
x_top = (-x_abs).argsort()[:n]
x2_top = (-x2_abs).argsort()[:n]
x3_top = (-x3_abs).argsort()[:n]
x4_top = (-x4_abs).argsort()[:n]
```

```python
np.std(x_top), np.std(x2_top), np.std(x3_top), np.std(x4_top)
```

Based on the duration of the audio file, the scale of the FFT is different. 
Using MinMaxScaler to preserve distribution but keep scale consistent. 

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
x_top_scaled = scaler.fit_transform(x_top.reshape(-1, 1))
x2_top_scaled = scaler.fit_transform(x2_top.reshape(-1, 1))
x3_top_scaled = scaler.fit_transform(x3_top.reshape(-1, 1))
x4_top_scaled = scaler.fit_transform(x4_top.reshape(-1, 1))

x_top_scaled = x_top_scaled.reshape(1, -1)
x2_top_scaled = x2_top_scaled.reshape(1, -1)
x3_top_scaled = x3_top_scaled.reshape(1, -1)
x4_top_scaled = x4_top_scaled.reshape(1, -1)
```

```python
x_top_scaled, x2_top_scaled, x3_top_scaled, x4_top_scaled
```

```python
np.std(x_top_scaled), np.std(x2_top_scaled), np.std(x3_top_scaled), np.std(x4_top_scaled)
```

Comparison is much better now.
